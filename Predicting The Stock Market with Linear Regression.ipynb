# Predicting The Stock Market with Linear Regression

A predictive model is developed using dataset extracted from S&P500 Index from 1950 to 2015.

The model will be trained with data from 1950-2012 and try to make predictions from 2013-2015.

The data set is in sphist.csv and the columns are:

- Date -- The date of the record.
- Open -- The opening price of the day (when trading starts).
- High -- The highest trade price during the day.
- Low -- The lowest trade price during the day.
- Close -- The closing price for the day (when trading is finished).
- Volume -- The number of shares traded.
- Adj Close -- The daily closing price, adjusted retroactively to include any corporate actions. 

First of all, it is necessary to conveert the data type of the Date column to datetime. That way it is easier to split the data.

```python
import datetime as datetime
import pandas as pd

data = pd.read_csv('sphist.csv')
```

## Creating new indicators

Datasets from stock market are different from most datasets in which each row is independent.
Stocks data is sequential and each observation comes a day after the previous observation. Thus, the observations are not all independent and you can't treat them as such.

It is very important not to inject future knowledge into past rows when training and predicting.

The time series nature of the data means that we can generate indicators to make our model more accurate.


Several new columns are generated that incorporate information from multiple prior rows into one and makes predictions much more accurate. These new indicators are:

- Prior 5 days average and standard deviation.

- Prior 30 days average and standard deviation.

- Prior 365 days average and standard deviation.

Since it is not a very large dataset the creation of the new columns is made iterating with iterrows().
It is also simpler to understand what is happening than by using the rolling function and shift.

```python
data = data.reset_index().drop('index', axis = 1)
data['5day_avg'] = 0.0
data['30day_avg'] = 0.0
data['365day_avg'] = 0.0

data['5day_std'] = 0.0
data['30day_std'] = 0.0
data['365day_std'] = 0.0

window_sizes = [5, 30, 365]


def window_avg_std(window_size):
    for index, row in data.iterrows():
        if index < window_size:
            data.loc[index, '{}day_avg'.format(window_size)] = 0
            data.loc[index, '{}day_std'.format(window_size)] = 0
            #row['5day_avg'] = 0
        else:        
            data.loc[index, '{}day_avg'.format(window_size)] = data['Close'][index-window_size:index].mean()
            data.loc[index, '{}day_std'.format(window_size)] = data['Close'][index-window_size:index].std()
```

```python
for window_size in window_sizes:
    window_avg_std(window_size)
```

The first 365 rows could be correctly generated because there was not historical data to create all the new indicators.

Therefore all the rows before 1951-06-19 should be deleted.

```python
data = data[data['Date'] > datetime.datetime(year = 1951, month = 6, day = 19)]
```

```python
data.head()
```

## Splitting the data

The training set is comprised between 1951 - 2013 while the test set has data belonging to 2013-2015.

```python
train = data[data['Date'] < datetime.datetime(year = 2013, month = 1, day = 1)]
test = data[data['Date'] > datetime.datetime(year = 2013, month = 1, day = 1)]
```

```python
train.head()
```

```python
test.head()
```

## Training and Predicting

The error metric is Mean Absolute Error as it is more intuitive than Mean Squared Errors when comparing to the real prices.

All the original columns (Close, High, Low, Open, Volume, Adj Close, Date) contain information around the future and they should not be used as features.

The features will be the new indicators: averages and standar deviations of previous Close values.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

features = ['5day_avg','30day_avg','365day_avg','5day_std','30day_std','365day_std']
target = 'Close'

model = LinearRegression()
model.fit(train[features], train[target])

predictions = model.predict(test[features])

mae = mean_absolute_error(test[target], predictions)

print(mae)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(15,10))
plt.plot(test['Date'], test['Close'], c = 'red')
plt.plot(test['Date'], predictions, c = 'blue')
plt.xticks(rotation='vertical')
plt.xlabel('Date', fontsize=13)
plt.ylabel('Closing Value', fontsize=13)
plt.legend(['Real','Predicted'],loc='lower right', fontsize='xx-large')
```
